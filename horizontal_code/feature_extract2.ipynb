```python
# d = pd.read_csv("../train/browse_history_train.txt",header=None,
#                     names=['用户标识','浏览时间','浏览行为数据','浏览子行为编号'])
d = pd.read_csv("/mnt/c/Users/cymle/Desktop/data/browse_history.csv")
d['time']=d['time']//86400
d
```

```python
gb=d.groupby(["uid"],as_index=False)

x1=gb['scan_record'].agg({'x_scan_recordsum' : 'sum','x_scan_recordmax' : 'max','x_scan_recordmean' : 'mean'
                     ,'x_scan_recordmin' : 'min','x_scan_recordstd' : 'std','x_scan_recordvar' : 'var'})
xx=gb['sub_scan_record'].apply(lambda x:np.unique(x).size)
x2=gb['sub_scan_record'].agg({'x_sub_scan_recordcount' : 'count'})
x2['x_sub_scan_recordcount_filter']=xx

feature = x1
feature=pd.merge(feature, x2,how='left', on = "uid")
```

```python
#统计放款前用户浏览子行为个类别统计信息

d1 = pd.get_dummies(d['sub_scan_record'], prefix='sub_scan_record')
d1 = d.join(d1)
d1.drop('sub_scan_record', 1)

gb=d1.groupby(["uid"],as_index=False)

x1=gb['sub_scan_record_1'].agg({'x_sub_scan_record_1' : 'sum'})
x2=gb['sub_scan_record_2'].agg({'x_sub_scan_record_2' : 'sum'})
x3=gb['sub_scan_record_3'].agg({'x_sub_scan_record_3' : 'sum'})
x4=gb['sub_scan_record_4'].agg({'x_sub_scan_record_4' : 'sum'})
x5=gb['sub_scan_record_5'].agg({'x_sub_scan_record_5' : 'sum'})
x6=gb['sub_scan_record_6'].agg({'x_sub_scan_record_6' : 'sum'})
x7=gb['sub_scan_record_7'].agg({'x_sub_scan_record_7' : 'sum'})
x8=gb['sub_scan_record_8'].agg({'x_sub_scan_record_8' : 'sum'})
x9=gb['sub_scan_record_9'].agg({'x_sub_scan_record_9' : 'sum'})
x10=gb['sub_scan_record_10'].agg({'x_sub_scan_record_10' : 'sum'})
x11=gb['sub_scan_record_11'].agg({'x_sub_scan_record_11' : 'sum'})

feature=pd.merge(feature, x1,how='left', on = "uid")
feature=pd.merge(feature, x2,how='left', on = "uid")
feature=pd.merge(feature, x3,how='left', on = "uid")
feature=pd.merge(feature, x4,how='left', on = "uid")
feature=pd.merge(feature, x5,how='left', on = "uid")
feature=pd.merge(feature, x6,how='left', on = "uid")
feature=pd.merge(feature, x7,how='left', on = "uid")
feature=pd.merge(feature, x8,how='left', on = "uid")
feature=pd.merge(feature, x9,how='left', on = "uid")
feature=pd.merge(feature, x10,how='left', on = "uid")
feature=pd.merge(feature, x11,how='left', on = "uid")
feature
```

```python
dataset = feature
temp=dataset[:]
#统计行和列的缺失值数
dataset=dataset.fillna(-1)
temp['user_no_data_count']=(dataset==-1).sum(axis=1)
#用均值填充缺失值
temp.fillna(temp.mean(),inplace=True)
temp
```
